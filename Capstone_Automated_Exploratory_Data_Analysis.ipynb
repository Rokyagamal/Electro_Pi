```python
# Import libraries numpy, pandas, matplotlib, seaborn for data analysis processes
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read DataFrame
data_Passengers = pd.read_csv('International_Report_Passengers.csv')
```

```python
# Check the first few rows of the DataFrame
data_Passengers.head()
```

```python
# Check the data types of each column
data_Passengers.dtypes
```

```python
# Check the summary statistics of the numerical columns
data_Passengers.describe()
```

```python
# Show that Month (3) 'March' in all years has the highest total flights over years (1990-2020)
data_Passengers['Month'].value_counts()
```

```python
# Show that Month (3) 'March' in all years has the highest total flights over years (1990-2020)
top_airlines = data_Passengers.groupby("Month")["Year"].sum().nlargest(10)
plt.plot(top_airlines.index, top_airlines.values)
plt.xlabel("Month")
plt.ylabel("Total Flights")
plt.title("Month with highest flights over years")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Show that Year (2018) has the highest total flights over years (1990-2020)
data_Passengers['Year'].value_counts()
```

```python
# Show that Year (2018) has the highest total flights over years (1990-2020)
top_airlines = data_Passengers.groupby("Year")["Month"].sum().nlargest(10)
plt.plot(top_airlines.index, top_airlines.values)
plt.xlabel("Year")
plt.ylabel("Total Flights")
plt.title("Month with highest flights over years")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Monthly trend of Total flights
# Show that the most number in flights was in July (7) Month
monthly_totals = data_Passengers.groupby(data_Passengers["Month"])["Total"].sum()
sns.lineplot(data=monthly_totals)
plt.xlabel("Year-Month")
plt.ylabel("Total Flights")
plt.title("Monthly Trend of Total Flights")
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
# Group the data by year and month and calculate the total for each month
monthly_totals = data_Passengers.groupby(["Year", "Month"])["Total"].sum().reset_index()
```

```python
# Find the highest month and corresponding year
highest_month = monthly_totals.loc[monthly_totals["Total"].idxmax()]
highest_month_year = highest_month["Year"]
```

```python
# Print the highest month and corresponding year
# Show that the most number in flights was in July (7) Month
# And was in year 2018
print("Highest Month in Total:")
print("Month:", highest_month["Month"])
print("Year:", highest_month_year)
```

```python
# Visualization (optional)
# Create a line plot of monthly totals
sns.lineplot(data=monthly_totals, x="Month", y="Total", hue="Year")
plt.xlabel("Month")
plt.ylabel("Total")
plt.title("Monthly Total")
plt.legend(title="Year")
plt.show()
```

```python
grouped_data = data_Passengers.groupby("carrier").sum()
```

```python
# Show the relation between column 'Scheduled' and 'Charter'
sns.histplot(data=data_Passengers, x="Scheduled", bins=20, color="blue", label="Scheduled")
sns.histplot(data=data_Passengers, x="Charter", bins=20, color="green", label="Charter")
plt.xlabel("Number of Flights")
plt.ylabel("Frequency")
plt.title("Distribution of Flights")
plt.legend()
plt.show()
```

```python
# Comparison of metrics between domestic and foreign carriers
sns.boxplot(data=data_Passengers, x="carriergroup", y="Scheduled")
plt.xlabel("Carrier Group")
plt.ylabel("Scheduled Flights")
plt.title("Comparison of Scheduled Flights between Domestic and Foreign Carriers")
plt.show()
```

```python
# Show the count of values of column 'carrier' that shows that 'AA' is the most 'Airlineid'
data_Passengers['carrier'].value_counts()
```

```python
# Top airlines by Total flights
# Show the count of values of column 'carrier' that shows that 'AA' is the most 'Airlineid'
top_airlines = data_Passengers.groupby("carrier")["Total"].sum().nlargest(10)
sns.barplot(x=top_airlines.index, y=top_airlines.values)
plt.xlabel("Airline")
plt.ylabel("Total Flights")
plt.title("Top 10 Airlines by Total Flights")
plt.xticks(rotation=45)
plt.show()
```

```python
# Top Foregin Gateway Airport Code by fg_apt_id and fg_apt
# Show 'YYZ' Gateway is the most Foreign Gateway Code acrosss years (1990-2020)
top_airlines = data_Passengers.groupby("fg_apt")["fg_apt_id"].sum().nlargest(10)
sns.barplot(x=top_airlines.index, y=top_airlines.values)
plt.xlabel("Foreign Gateway Code")
plt.ylabel("Total Flights")
plt.title("Top 10 Foreign Gateway Code by Total Flights")
plt.xticks(rotation=45)
plt.show()
```

```python
# Top US Gateway Airport Code by usg_apt_id and usg_apt
# Show 'MIA' Gateway is the most US Gateway Code acrosss years (1990-2020)
top_airlines = data_Passengers.groupby("usg_apt")["usg_apt_id"].sum().nlargest(10)
sns.barplot(x=top_airlines.index, y=top_airlines.values)
plt.xlabel("US Gateway Code")
plt.ylabel("Total Flights")
plt.title("Top 10 US Gateway Code by Total Flights")
plt.xticks(rotation=45)
plt.show()
```

```python
# Top Foregin Gateway World Area Code by fg_wac and carrier
# Show 'AA' Gateway is the most Foreign Gateway Code acrosss years (1990-2020)
top_airlines = data_Passengers.groupby("carrier")["fg_wac"].sum().nlargest(10)
sns.barplot(x=top_airlines.index, y=top_airlines.values)
plt.xlabel("Foregin Gateway World Area Code")
plt.ylabel("Total Flights")
plt.title("Top 10 Foregin Gateway World Area Code across Total Flights")
plt.xticks(rotation=45)
plt.show()
```

```python
# Top US Gateway World Area Code by usg_wac and carrier
# Show 'AA' Gateway is the most US Gateway Code acrosss years (1990-2020)
top_airlines = data_Passengers.groupby("carrier")["usg_wac"].sum().nlargest(10)
sns.barplot(x=top_airlines.index, y=top_airlines.values)
plt.xlabel("US Gateway World Area Code")
plt.ylabel("Total Flights")
plt.title("Top 10 US Gateway World Area Code across Total Flights")
plt.xticks(rotation=45)
plt.show()
```
