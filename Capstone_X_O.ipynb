```python
# Function to print the game board
def print_board(board):
    for row in board:
        print("|".join(row))
        print("-" * 5)
```

```python
# Function to check for a win
def check_win(board, player):
    # Check rows
    for row in board:
        if all(square == player for square in row):
            return True

    # Check columns
    for col in range(3):
        if all(board[row][col] == player for row in range(3)):
            return True

    # Check diagonals
    if board[0][0] == board[1][1] == board[2][2] == player:
        return True

    if board[0][2] == board[1][1] == board[2][0] == player:
        return True

    return False
```

```python
# Function to check for a tie
def check_tie(board):
    for row in board:
        if "" in row:
            return False
    return True
```

```python
# Function to ask the player to choose "X" or "O"
def choose_marker():
    marker = input("Choose 'X' or 'O': ").upper()
    while marker not in ["X", "O"]:
        marker = input("Invalid choice! Choose 'X' or 'O': ").upper()
    return marker
```

```python
# Function to ask a player to play again
def play_again():
    choice = input("Play again? (yes/no): ").lower()
    while choice not in ["yes", "no"]:
        choice = input("Invalid choice! Play again? (yes/no): ").lower()
    return choice == "yes"
```

```python
# Create an empty game board
board = [["", "", ""], ["", "", ""], ["", "", ""]]
```

```python
# Ask the player to choose a marker
current_player = choose_marker()

while True:
    # Print the current state of the game board
    print_board(board)

    # Prompt the current player for their move
    row = int(input("Enter the row (0-2): "))
    col = int(input("Enter the column (0-2): "))

    # Validate the position
    if row not in range(3) or col not in range(3):
        print("Invalid position! Try again.")
        continue

    # Update the game board with the player's move
    if board[row][col] == "":
        board[row][col] = current_player
    else:
        print("Position already taken! Try again.")
        continue

    # Check if the current player has won
    if check_win(board, current_player):
        print_board(board)
        print("Player", current_player, "wins!")
        if not play_again():
            break
        else:
            # Reset the game board and choose new markers
            board = [["", "", ""], ["", "", ""], ["", "", ""]]
            current_player = choose_marker()
            continue

    # Check if the game is tied
    if check_tie(board):
        print_board(board)
        print("It's a tie!")
        if not play_again():
            break
        else:
            # Reset the game board and choose new markers
            board = [["", "", ""], ["", "", ""], ["", "", ""]]
            current_player = choose_marker()
            continue

    # Switch to the other player
    current_player = "O" if current_player == "X" else "X"
```
